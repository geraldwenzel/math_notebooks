Graph $sin(x) < cos(y)$

```python
x_coords = np.linspace(-9.4, 9.4, 200)
y_max=6.2
y_coords = np.linspace(-y_max, y_max, 200)

for _x in x_coords:
    for _y in y_coords:
        if sin(_x) < cos(_y):
            plt.plot(_x, _y, 'ro')
            
plt.axis(ymin=-y_max, ymax=y_max, xmin=-9.4, xmax=9.4)
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.axhline(0, color='black', lw=2)
_=plt.axvline(0, color='black', lw=2)
```
